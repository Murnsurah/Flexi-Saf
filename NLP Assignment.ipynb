## Lower casing
This is to convert every letter in the column to lower case

```python
df['text_lower'] = df["text"].str.lower()
df.head()
```

## Removal of Punctuations
This is a text standardization process used for uniformity

```python
def remove_punctuation(text):
    """Removing punctuation using regular expressions"""
    return re.sub(r'[^\w\s]', '', text)

df["new_text"] = df["text"].apply(lambda text: remove_punctuation(text))
df.head()
```

```python
#removal of stop words
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english')
df["text_wo_stop"] = df["new_text"].apply(lambda text: " ".join(vectorizer.build_tokenizer()(text)))
df.head()
```
